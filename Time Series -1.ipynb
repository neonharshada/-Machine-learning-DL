```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'..\datafiles\bitcoindata.csv', parse_dates = True, index_col='Date')
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(14,4))
plt.plot(df['Open'], marker='*', ms=12, mfc='r')
plt.show()
```

```python
df['Day'] = df.index.day
df['Month'] = df.index.month
df['Year'] = df.index.year
df['WkDays'] = df.index.weekday
# 0 is Monday
```

```python
df
```

```python
df.boxplot(column='Open', by='WkDays', figsize=(18,4))
plt.show()
```

```python
df['MA5'] = df['Open'].rolling(5).mean()
```

```python
plt.figure(figsize=(14,4))
plt.plot(df['Open'], marker='*', ms=12, mfc='r')
plt.plot(df['MA5'] , marker='H', ms=12, mfc='g')
plt.show()
```
