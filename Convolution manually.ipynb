```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
img = image.load_img("parrot.jpg", color_mode = 'grayscale')
plt.imshow(img, cmap='gray');
```

```python
filter1 = np.array([[-1, 1, -1],
                              [-1, 1, -1],
                              [-1, 1, -1]])
```

```python
img_arr = np.asarray(img)
type(img_arr)
img_arr.shape
```

```python
new_img = np.zeros((178, 279))
plt.imshow(new_img, cmap='gray');
```

```python
for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
        new_img[i,j] = np.tensordot(img_arr[i:i+3, j:j+3],filter1)
```

```python
fig,ax = plt.subplots(1,2, figsize=(8,10))
ax[0].imshow(img, cmap='gray');
ax[1].imshow(new_img, cmap='gray');
```

```python
Gx = np.array([[1, 0, -1],
                         [2, 0, -2],
                         [1, 0, -1]])
Gy = Gx.T
```

```python
Gy
```

```python
for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
        x= np.tensordot(img_arr[i:i+3, j:j+3],Gx)
        y= np.tensordot(img_arr[i:i+3, j:j+3],Gy)
        new_img[i,j] = np.sqrt(x**2+ y**2)
```

```python
fig,ax = plt.subplots(1,2, figsize=(8,10))
ax[0].imshow(img, cmap='gray');
ax[1].imshow(new_img, cmap='gray');
```
