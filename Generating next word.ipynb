```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
```

```python
text = ['hello how are you I hope you are doing good', 
       'good morning students I hope you had a good night sleep',
       'you need to take it seriously otherwise you will fail']
```

```python
# lowercase all words and create a vocabulary
tokens=[]
for eng in text:
    for word in eng.split():
        if word.lower() not in tokens:
            tokens.append(word.lower())
tokens
```

```python
lines = []
for i in range(5,len(tokens)):
    seq = tokens[i-5:i]
    #print(seq)
    line = ' '.join(seq)
    lines.append(line)

lines
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(lines)
#creating of dictionary 
#mapping  - words to number
tokenizer.word_index
```

```python
#converting words to numbers
sequences = tokenizer.texts_to_sequences(lines)
```

```python
#converting words to numbers
sequences = tokenizer.texts_to_sequences(lines)
sequences
```

```python
sequences = np.array(sequences)
```

```python
X = sequences[:,:-1]
y = sequences[:,-1]
print(X.shape, y.shape)
```

```python
vocab_size = len(tokenizer.word_index)+1
```

```python
y = to_categorical(y, num_classes=vocab_size)
print(y.shape)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 5, input_length=4))
model.add(LSTM(3))
model.add(Dense(vocab_size, activation='softmax'))
model.summary()
model.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
model.fit(X, y, epochs=1000, batch_size=2)
```

```python
test= ['hope sleep good how']
test = np.array(tokenizer.texts_to_sequences(test)).reshape(1,4)
```

```python
np.argmax(model.predict(test))
```
