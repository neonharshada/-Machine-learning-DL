```python
X = data.iloc[:,1:-2].values
y = data.iloc[:,-1].values
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
# Split the data into training and testing sets 
#stratify = y will do a stratified sampling
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                test_size = 0.2, 
                                                random_state = 0, stratify=y) 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=10)
# n_estimators = no of trees
model.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, model.predict(X_test));
```

```python
print(classification_report(y_test, model.predict(X_test)))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_sc = sc.fit_transform(data.iloc[:,1:-1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sc, y, 
                                                test_size = 0.2, 
                                                random_state = 0, stratify=y) 
```

```python
model = RandomForestClassifier(n_estimators=10)
# n_estimators = no of trees
model.fit(X_train, y_train).score(X_test, y_test)
```

```python
confusion_matrix(y_test,model.predict(X_test))
```

```python
print(classification_report(y_test, model.predict(X_test)))
```

```python
### Unsupervised Model
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(n_estimators=100)
clf.fit(X_train)
y_pred = clf.predict(X_test)
# -1 if data point is outlier, 1 if data point is normal
```

```python
y_pred
```

```python
# In our case 1 label is for fraud and 0 label is for normal
# so to compare, we will replace 1 with 0 and -1 with 1 
```

```python
y_pred[y_pred==1] = 0
y_pred[y_pred==-1] = 1
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
